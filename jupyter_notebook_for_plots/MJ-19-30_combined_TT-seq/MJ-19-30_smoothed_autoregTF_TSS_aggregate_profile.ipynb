```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy as sc
import math

plots_dir = 'S:\\lab_winter\\martin\\MED14_PRO-seq_final\\plots\\' # directory for output plots
```

```python
############################
#### load in raw values ####
############################
## raw values generated with computeMatrix and plotHeatmap, but manually smoothed (running mean of +/- 3 bins) for the autoregTFs due to spiky signal (only 24 genes)

plot_data = pd.read_csv('S:\\lab_winter\\martin\\MED14_PRO-seq_final\\plots\\enrichment_heatmaps\\MJ-19-30_TSS_autoregTF_6cell-smoothed_profile_values.txt', sep="\t", index_col=0).T
```

```python
###########################
#### make the lineplot ####
###########################

plt.close('all')
sns.set_style('ticks')
sns.set_context('talk')
plt.figure(figsize=(5,5))

# plot
plt.plot(plot_data.index, plot_data.iloc[:,0], color="black")
plt.plot(plot_data.index, plot_data.iloc[:,1], color="grey")
plt.plot(plot_data.index, plot_data.iloc[:,2], color="darkred")
plt.plot(plot_data.index, plot_data.iloc[:,3], color="orange")
plt.fill_between(plot_data.index,plot_data.iloc[:,0], alpha=0.5, color="black", linewidth=0)
plt.fill_between(plot_data.index,plot_data.iloc[:,1], alpha=0.5, color="grey", linewidth=0)
plt.fill_between(plot_data.index,plot_data.iloc[:,2], alpha=0.5, color="darkred", linewidth=0)
plt.fill_between(plot_data.index,plot_data.iloc[:,3], alpha=0.5, color="orange", linewidth=0)

# format
plt.legend(loc=2, fontsize=10)
plt.xlim(0,99)
plt.ylim(0,0.7)
plt.xticks([0,49,99], ['-0.5', 'TSS', '0.5 kb'])
plt.yticks(np.arange(0,0.6,0.25))

plt.savefig(plots_dir + 'enrichment_heatmaps\\MJ-19-30_aggregate_coverage_deeptools_manually-smoothed_24autoregTFs.pdf', dpi=300)
plt.savefig(plots_dir + 'enrichment_heatmaps\\MJ-19-30_aggregate_coverage_deeptools_manually-smoothed_24autoregTFs.png', dpi=300)
#plt.show()
```
